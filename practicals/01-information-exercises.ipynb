```python
# Make HTTP requests for internet reqources.
import urllib.request

# The URL of a text version of Alice in Wonderland.
book_url = 'https://www.gutenberg.org/files/11/11-0.txt'

# Get the book.
book = list(urllib.request.urlopen(book_url))

# Decode the lines and strip line endings.
book = [line.decode('utf-8-sig').strip() for line in book]

# Get a sample paragraph - I looked for this by hand.
paragraph = ' '.join(book[58:63])

# Show the paragraph.
print(paragraph)
```

```python
# Let's lower-case it.
alice = paragraph.lower()

# All letters and a space.
chars = 'abcdefghijklmnopqrstuvwxyz '

# And strip anything that is not a letter or space.
alice = ''.join([c for c in alice if c in chars])

# Show the paragraph now.
print(alice)
```

```python
# For doing (pseudo-)random things in Python.
import random

# Print a random character, for example.
print(random.choice(chars))
```

```python
# Get the length of alice.
N = len(alice)

# Generate N random characters from chars.
gener = random.choices(chars, k=N)

# Join them together in a string.
gener = ''.join(gener)

# Print.
print(gener)
```

```python
# Get the whole book in one big string.
sbook = ''.join(book[26:]).lower()

# Create the weights - count the occurences of each character in the whole book.
weights = [sbook.count(c) for c in chars]

# Show the weights.
weights
```

```python
# Generate a string using those weights.
wgenr = random.choices(chars, weights=weights, k=N)

# Join them together in a string.
wgenr = ''.join(wgenr)

# Print.
print(wgenr)
```

### Explain why the log of zero is undefined

We can start with division by zero being undefined.<br>
Firstly, consider the case where we try and divide 1 by very small numbers:<br>


```python
a = (1 / 0.1)
b = (1 / 0.0000001)
c = (1 / 0.0000000001)
a, b, c
```

It can be seen that as the numbers get infinitessimally small, the result approaches infinity.
<br>
<br>
Now, what if we try and divide 1 by very small negative numbers?

```python
c = (1 / -0.1)
d = (1 / -0.0000001)
e = (1 / -0.0000000001)
c, d, e
```

It can be seen, therefore, that as the negative numbers get infinitessimally small,<br>
the result approaches negative infinity.
So there's two potential outcomes:<br>
1) If you try and divide 1 by very small positive numbers, the result approaches infinity.
2) If you try and divide 1 by very small negative numbers, the result approaches negative infinity.

So, 1 / 0 gives two answers: infinity and negative infinity. 
Nothing can be further apart then infinity and negative infinity and therefore, division by zero has to be undefined.

y = log<sub>b</sub>(x) ⟺ b<sup>y</sup> = x<br>

Where:

y = 2<sup>x</sup>


```python
import math
import matplotlib.pyplot as plt
import numpy as np
 
# Create input range of numbers in form of an array.
x = np.linspace(-9, 9, 100)
y1 = 2 ** x
# The below gives a runtime warning, so was omitted. If x contains a positive
# range, the error goes
#y2 = np.log2(x)
fig = plt.figure(figsize = (14, 7))
# Create the plot
plt.plot(x, y1, alpha = 0.4, label ='y = $2^{x}$',
         color ='green', linestyle ='dotted',
         linewidth = 2, marker ='D',
         markersize = 3, markerfacecolor ='red',
         markeredgecolor ='blue')
 
# Add a title
plt.title('y = $2^{x}$')
 
# Add X and y Label
plt.xlabel('x axis')
plt.ylabel('y axis')

# Add a grid
plt.grid(alpha =.6, linestyle ='--')
 
# Add a Legend
plt.legend()
 
# Show the plot
plt.show()
```

There's no number that can be substituted for x to make 2<sup>x</sup> = 0.

As we don't know what that number is, we can't define it - therefore, it's classed as undefined.
