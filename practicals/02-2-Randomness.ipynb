The first thing we need to do is create a random number generator and seed it with a number of our choice. This can then be used across all distributions to have some type of consistency.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create new, seeded random number generator.
seed = 3
rng = np.random.default_rng(seed)
rng
```

```python
# Use the random number generator to generate an array of numbers
#
sample_size = 1000
samples = rng.standard_exponential(sample_size)
```

### Standard Exponential Distribution<br>
The information for the standard exponential distribution can be found [here](https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.standard_exponential.html#numpy.random.Generator.standard_exponential).

Exponential decay can be used to show the value of something decreasing over a range, for example:<br>
- Radioactive decay
- Rate of infections in a population

```python
# Used the random number generated 
samples = rng.standard_exponential(sample_size)

# Specify plotting options and display the plot
plt.hist(samples, bins = 100, density = True)
plt.show()
```

### Weibull Distribution:


More information can be found on the Weibull distribution [here](https://statisticsbyjim.com/probability/weibull-distribution/).<br>

The Weibull distribution is typically used for:<br>
- Reliability analysis (for assessing time to failure for systems/parts)
- Engineering

```python
a = 5 # shape
w_samples = rng.weibull(a, sample_size)
#w_samples
```

```python
x = np.arange(1,100.)/50.
def weib(x,n,a):
    return (a / n) * (x / n)**(a - 1) * np.exp(-(x / n)**a)
```

```python
count, bins, ignored = plt.hist(rng.weibull(a,1000))
x = np.arange(1,100.)/50.
scale = count.max()/weib(x, 1., a).max()
plt.plot(x, weib(x, 1., a)*scale)
plt.show()
```

So far, it looks pretty symetrical and has a normal curve shape.
If the shape is changed, we can see different outputs.<br>

Set the shape to 1:

```python
a1 = 1 # shape
w_samples = rng.weibull(a1, sample_size)
```

```python
x = np.arange(1,100.)/50.
def weib(x,n,a1):
    return (a1 / n) * (x / n)**(a1 - 1) * np.exp(-(x / n)**a1)
```

```python
count, bins, ignored = plt.hist(rng.weibull(a1,sample_size))
x = np.arange(1,100.)/50.
scale = count.max()/weib(x, 1., a1).max()
plt.plot(x, weib(x, 1., a1)*scale)
plt.show()
```

So the distribution decreases exponentially when the shape is 1. If this was looking at failure rates, we can say that the failure rate over time is consistent.
